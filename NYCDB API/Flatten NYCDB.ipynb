```python
reg_dfs = []
for index, row in data.iterrows():
    bbl = row['code']
    
    api_str = Template(nycdb_url['registration']).safe_substitute({ 'bbl': bbl})
    if (api_str not in cached_data):        
        reg_df = fetch_and_read_data_into_df(api_str)
        cached_data[api_str] = reg_df
        
    reg_dfs.append(reg_df)
```
